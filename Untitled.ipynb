Analiza receptov mesnih jedi

```python
import pandas as pd
%matplotlib inline
jedi = pd.read_csv('jedi.csv')
pd.options.display.max_rows = 20
```

Najprej si oglejmo podatke.

```python
jedi
```

Kako popularna je spletna stran v zadnjih letih?

```python
jedi_po_letih = jedi.groupby('leto')
```

```python
jedi_po_letih.size()
```

```python
jedi_po_letih.size().plot(figsize=(10,5), title = 'Število receptov po letih')
```

Opazimo, da je pred 15 leti število receptov močno naraščalo, v zadnjih letih pa se zmanjšuje. 
Ali je to odvisno od avtorjev? Poglejmo, kdo je prispeval največ receptov.

```python
jedi['avtor'].value_counts().head(10)
```

Marinka je zmagovalka, zato si podrobneje oglejmo njene recepte.

```python
jedi[jedi['avtor'] == 'Marinka'].mean()
```

To je njen povprečen recept. Kasneje se bomo vračali k njemu za primerjavo.

```python
jedi[jedi['avtor'] == 'Marinka'].sort_values('leto')
```

Vidimo, da je objavljala med letoma 1997 in 2007, kar se ujema z leti z največ objavami.

Sedaj si oglejmo razmerje med zastopanostjo vrst jedi.

```python
jedi_po_vrsti = jedi.groupby('vrsta')
jedi_po_vrsti.size().plot.pie(figsize=(8, 8), subplots=True, autopct='%.2f', fontsize=10, title = 'Zastopanost jedi po vrstah')
```

Kot vidimo, je najbolj priljubljena perutnina, notranji deli, drobnica in koline pa najmanj. Je morda vzrok v zahtevnosti priprave?

```python
jedi.sort_values('st. porcij', ascending=False).head(20).plot.bar(x='jed', y='st. porcij')
```
